```
P = (10,20,30)
Qd = (200,150,100)
Qo = (100,150,200)
 
bo = (Qo[1]-Qo[0])/(P[1]-P[0])
ao = Qo[0]-bo*P[0]
Co = (ao,bo)
Co
print(Co)
print("La ecuaci贸n de la oferta es:", "Qo = ", ao, "+", bo, "*P")
 
import matplotlib.pyplot as plt
plt.plot(Qo,P)
 
bd = (Qd[1]-Qd[0])/(P[1]-P[0])
ad = Qd[0]-bd*P[0]
Cd = (ad,bd)
print(Cd)
print("La ecuaci贸n de la Demanda es:", "Qd = ", ad, "-", -bd, "*P")
 
plt.plot(Qd,P)
 
plt.plot(Qo,P)
plt.plot(Qd,P)
 
Pe = (ad-ao)/(bo-bd)
Qe = ao+bo*Pe
E = (Pe,Qe)
print("El equilibrio es", E)
 
lrPe = (Pe,Pe,Pe)
lrQe = (Qe,Qe,Qe)
 
plt.plot(Qo,lrPe)
plt.plot(lrQe,P)
```
