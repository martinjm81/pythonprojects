```python
I0 = 200
C0 = 400
G0 = 150 
Y0_C = I0 + C0 + G0
print(Y0_C)
```

```python
Y0_C = I0 + C0 + G0
print(Y0_C)
CA = 75
c = 0.9
C0_E1 = CA + c*Y0_C
Y0_E1 = I0 + C0_E1 + G0
print(C0_E1, Y0_E1)
C0_E2 = CA + c*Y0_E1
Y0_E2 = I0 + C0_E2 + G0
print(C0_E2,Y0_E2)
C0_E3 = CA + c*Y0_E2
Y0_E3 = I0 + C0_E3 + G0
print(C0_E3,Y0_E3)
```

```python
C0_E = [0]*100
Y0_E = [0]*100
C0_E[0] = CA + c*Y0_C
Y0_E[0] = I0 + C0_E[0] + G0


for i in range(1, 100):
    j = i - 1
    C0_E[i] = CA + c*Y0_E[j]    
    Y0_E[i] = I0 + C0_E[i] + G0
print(Y0_E)
```

```python
print('\n'.join(str(y) for y in Y0_E))
```

```python
m = 1/(1-c)
print(m)
Y0_Em = m*(CA+I0+G0)
print(Y0_Em)
```
