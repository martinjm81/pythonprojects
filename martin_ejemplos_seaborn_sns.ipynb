# Datos Sintéticos

```python
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.DataFrame({'var1': np.random.rand(100),'var2': np.random.rand(100),
    'var3': np.random.rand(100), 'var4': np.random.rand(100)
})

sns.set_theme() #(opcional)
```

# 1) Histplot

```python
#histogramas apilados
sns.histplot(data, kde=True)
```

```python
sns.displot(data, kde=True)
```

# 2) Kernel Density Plot

```python
#densidadesdekernel
sns.kdeplot(data)
```

```python
#combinaciones
import matplotlib.pyplot as plt

variables = ['var1', 'var2', 'var3', 'var4']

fig, axes = plt.subplots(len(variables), len(variables), figsize=(15, 15))

for i, var_i in enumerate(variables):
    for j, var_j in enumerate(variables):
        if i == j:
            sns.histplot(data[var_i], ax=axes[i, j], kde=True)
        else:
            sns.kdeplot(x=data[var_j], y=data[var_i], ax=axes[i, j], cmap="Blues")

plt.tight_layout()
plt.show()
```

# 3) Scatter plot

```python
#dispersión
sns.scatterplot(data)
```

# 4) PairPlot

```python
#dispersión e histogramas
sns.pairplot(data)
```

```python
sns.pairplot(data, kind="reg")
```

```python
sns.pairplot(data, kind="reg", diag_kind="kde")
```

```python
sns.pairplot(data, diag_kind="kde")
```

```python
g = sns.pairplot(data, diag_kind="hist", corner=False)

for ax in g.diag_axes:
    for data_key, color in zip(data.columns, sns.color_palette()):
        sns.kdeplot(data[data_key], ax=ax, color=color)

plt.show()
```

```python
g = sns.pairplot(data, kind="reg", diag_kind="hist")

for ax in g.diag_axes:
    for data_key, color in zip(data.columns, sns.color_palette()):
        sns.kdeplot(data[data_key], ax=ax, color=color)
```

```python
g = sns.pairplot(data, kind="reg", diag_kind="hist",
                 plot_kws={'line_kws':{'color':'red'}})
plt.show()
```

# 5) Relplot

```python
sns.relplot(data)
```

```python
sns.relplot(data, kind="line")
```

# 6) JointPlot


```python
#dispersión y densidades
sns.jointplot(data)
```

```python
#distribucion conjunta
sns.jointplot(x='var1', y='var2', data=data, kind="reg")
sns.jointplot(x='var3', y='var4', data=data, kind="scatter")
```

```python
#distribucion conjuta
data['producto'] = data['var1'] * data['var2']
#sns.jointplot(x='producto', y='var3', data=data, kind="reg")
sns.jointplot(x='producto', y='var3', data=data, kind="hist")
plt.show()
```
